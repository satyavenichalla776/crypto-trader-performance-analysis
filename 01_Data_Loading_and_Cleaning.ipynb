# Bitcoin Market Sentiment Analysis & Price Prediction

## Project Objective
The objective of this project is to predict Bitcoin execution prices using market trading data and sentiment analysis. This project includes data cleaning, exploratory data analysis, feature engineering, machine learning model development, evaluation, and deployment preparation.

```python
#Import Libraries
# Ignore warning messages
import warnings
warnings.filterwarnings("ignore")
# Data manipulation
import pandas as pd
import numpy as np
# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
# Display graphs inside the notebook
%matplotlib inline
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()
missing_values
```

```python
# Calculate the percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.sort_values(ascending=False)
```

```python
# Count duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Duplicate Rows: {duplicate_rows}")
```

```python
# Remove duplicate records
bitcoin_data = df.drop_duplicates()
print("Duplicate rows removed successfully.")
```

```python
# Store the original number of rows
rows_before_cleaning = len(bitcoin_data)
# Remove rows containing missing values
bitcoin_data = bitcoin_data.dropna()
# Store the remaining number of rows
rows_after_cleaning = len(bitcoin_data)
print(f"Rows before cleaning : {rows_before_cleaning}")
print(f"Rows after cleaning  : {rows_after_cleaning}")
print(f"Rows removed         : {rows_before_cleaning - rows_after_cleaning}")
```

```python
# Ensure there are no missing values
bitcoin_data.isnull().sum()
```

```python
# Display the data type of each column
bitcoin_data.dtypes
```

```python
# Display unique values for every column
for column in bitcoin_data.columns:
    print(f"{column}: {bitcoin_data[column].nunique()} unique values")
```

```python
# Save the cleaned dataset for further analysis
bitcoin_data.to_csv("cleaned_bitcoin_data.csv", index=False)
print("Cleaned dataset saved successfully.")
```
